```python
import pandas as pd


df = pd.read_csv('Software_position.csv')
df
```

```python
import re

intern_position = r'Intern'
fresher_position = r'Fresher'
junior_position = r'Junior'
senior_position = r'Senior'

list_intern = []
list_fresher = []
list_junior = []
list_senior = []
list_other = []

for i in df["Position"]:
    if re.search(fresher_position, i, re.IGNORECASE):
        list_fresher.append(i)
    if re.search(intern_position, i, re.IGNORECASE):
        list_intern.append(i)
    if re.search(junior_position, i , re.IGNORECASE):
        list_junior.append(i)
    if re.search(senior_position, i, re.IGNORECASE):
        list_senior.append(i)
    else:
        list_other.append(i)

def count(lists):
    count = 0
    for i in lists:
        count += 1
    return count


def percent(position):
    return round(( position / df.shape[0] ) * 100)  

print()
print(count(list_other))
```

```python
import matplotlib.pyplot as plt
import numpy as np


y = np.array([count(list_fresher), count(list_intern), count(list_junior), count(list_senior), count(list_other)])
mylabels = [ str(count(list_fresher)), str(count(list_intern)), 
            str(count(list_junior)), str(count(list_senior))  
            , str(count(list_other))]
myexplode = [0, 0, 0, 0.3]

plt.pie(y, labels = mylabels)
plt.legend(title = "Number of each level:")
plt.show() 
```
