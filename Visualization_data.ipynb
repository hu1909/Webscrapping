```python
def visualize_data(size, title):
    # create data
    names = ['Intern', 'Fresher', 'Junior', 'Middle', 'Senior', 'Manager/Lead', 'Available for all level']
 
# Create a circle at the center of the plot
    my_circle = plt.Circle( (0,0), 0.7, color='white')

# Custom wedges
    plt.pie(size, labels=names, wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' })
    p = plt.gcf()
    p.gca().add_artist(my_circle)
    plt.title(title)
    plt.show()
    

def check_level(list_data, titles):
    list_intern = []
    list_fresher = []
    list_senior = []
    list_middle = []
    list_junior = []
    list_higher_position = []
    other = []
    for i in list_data:
        if re.search('Intern', i, re.IGNORECASE):
            list_intern.append(i)
        elif re.search('Fresher', i, re.IGNORECASE) or re.search('Graduate', i, re.IGNORECASE):
            list_fresher.append(i)
        elif re.search('Senior', i, re.IGNORECASE):
            list_senior.append(i)
        elif re.search('Middle', i, re.IGNORECASE) or re.search('Mid', i, re.IGNORECASE):
            list_middle.append(i)
        elif re.search('Junior', i, re.IGNORECASE):
            list_junior.append(i)
        elif re.search('Lead', i, re.IGNORECASE) or re.search('Manager', i, re.IGNORECASE) or re.search('CTO',i, re.IGNORECASE)or re.search('VP',i , re.IGNORECASE):
            list_higher_position.append(i)
        else:
            other.append(i)
    sizes = [len(list_intern), len(list_fresher), len(list_junior), len(list_middle), len(list_senior), len(list_higher_position), len(other)]
    visualize_data(sizes, titles)
    
       
check_level(list_software_jobs, 'All level of Software aspect')
check_level(list_data_jobs, 'All level of Data aspect')
check_level(list_hardware_jobs, 'All level of Hardware aspect')
check_level(list_security_jobs, 'All level of Security aspect')
```
